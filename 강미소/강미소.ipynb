```python
import ollama
def chat_with_llama(prompt):
    response = ollama.chat(model='llama3.2', messages=[
        {
            'role': 'user',
            'content': prompt,
        },
    ])
    return response['message']['content']

prompt= """
How are you today."""

guide = """
You are my friend. Talk with me.
"""

prompt = guide + prompt + "\n 답변 : "

output = chat_with_llama(prompt)
print(output)
```

```python
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

template_text = "안녕하세요, 제 이름은 {name} 이고, 나이는 {age}살입니다."
prompt_template = ChatPromptTemplate.from_template(template_text)

combined_prompt = (
            prompt_template
            + ChatPromptTemplate.from_template("\n\n밥을 좋아해요.")
            +"\n\n{language}로 번역해주세요."
)

output = combined_prompt.format(name="김철수", age=30, language="영어")

llm = ChatOllama(mode1="11ama3.2")
chain = combined_prompt|llm|StrOutputParser()
output_2 = chain.invoke({"age":30, "language":"영어", "name":"김철수"})
print(output_2)
```

```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain_core.prompts import ChatPromptTemplate

chat_prompt = ChatPromptTemplate.from_messages(
    [
	    SystemMessagePromptTemplate.from_template("이 시스템은 뇌과학 질문에 답변할 수 있습니다."),
	    HumanMessagePromptTemplate.from_template("{user_input}"),
]
)

messages = chat_prompt.format_messages(user_input="교감 신경계는 언제 활성화 되나요?")

from langchain_core.output_parsers import StrOutputParser
from langchain_community.chat_models import ChatOllama

llm = ChatOllama(model1="llama3.2")
chain = chat_prompt | llm | StrOutputParser()

print(chain.invoke({"user_input": "교감 신경계는 언제 활성화 되나요?"}))
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(
    file_path='cp1c.csv', 
    encoding='utf-8',
    source_column="메뉴"  
)

try:
    docs = loader.load()
    print(docs[0].page_content) 
except FileNotFoundError:
    print("파일을 찾을 수 없습니다. 경로를 확인하세요.")
except UnicodeDecodeError:
    print("파일 인코딩에 문제가 있습니다. 다른 인코딩을 시도해보세요.")
except KeyError as e:
    print(f"지정한 열 '{e}'이(가) CSV 파일에 존재하지 않습니다.")
except ValueError as e:
    print(e)
except IndexError:
    print("문서가 존재하지 않습니다. docs 리스트의 길이를 확인하세요.")
except Exception as e:
    print(f"예상치 못한 오류가 발생했습니다: {e}")

print(f"총 문서 수: {len(docs)}")
```

```python
from langchain_community.document_loaders import WebBaseLoader

url = 'https://ko.wikipedia.org/wiki/%EA%B0%9C'
loader = WebBaseLoader(url)

docs = loader.load()

print(len(docs))
print(len(docs[0].page_content))
print(docs[0].page_content)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(docs)

print(len(splits))
print(splits[10])
```

```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(
    model_name='BAAI/bge-m3',
    model_kwargs={'device':'cpu'}, 
    encode_kwargs={'normalize_embeddings':True},
)

from langchain_community.vectorstores import Chroma
vectorstore = Chroma.from_documents(splits, embeddings)

docs = vectorstore.similarity_search("청각에 대해서 설명해주세요.")
print(len(docs))
print(docs[0].page_content)
```
